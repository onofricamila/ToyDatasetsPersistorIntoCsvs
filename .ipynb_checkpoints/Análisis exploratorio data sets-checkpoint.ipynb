```python
def plot(x, y, title=''):
    plt.scatter(x, y, alpha=0.5)
    plt.xlabel('latitude')
    plt.ylabel('longitude')
    plt.ylim(10.0273686052, 23.9033785336)
    plt.xlim(54.3617373725, 69.1062472602)
    plt.title(title) 
    plt.show()
```

## Distribución de puntos en subsets (Sweden)

```python
t = [50000, 100000, 500000, 1000000]
```

```python
df.head(1)['to_timestamp'][1] # test
```

```python
def plot_subsets(is_incremental):
    t_i = 0
    i_ant = 1
    f_ant = df.head(1)['to_timestamp'][1]
    for i, r in df.iterrows():
        up_limit = t[t_i]
        lo_limit = 0 if (is_incremental) or (i == 1) else i_ant

        if i % up_limit == 0:
            x = df.latitude[lo_limit:i]
            y = df.longitude[lo_limit:i]
            plot(x, y, title = 'elementos del {} ({}) al {} ({})'.format(lo_limit, f_ant, i, r['to_timestamp']))

            i_ant = i
            f_ant = r['to_timestamp']
            
            if t_i != len(t)-1:
                t_i += 1
```

### Incremental

```python
plot_subsets(True)
```

### Distribución de puntos: ventanas (Sweden)

```python
plot_subsets(False)
```
